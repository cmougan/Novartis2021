```python
import pandas as pd
```

```python
split = pd.read_csv("../data/split.csv")
split = split[['region', 'validation']].drop_duplicates().reset_index(drop=True)
```

```python
split
```

# read all data

```python
from utils import get_data_from_path
```

```python
data = get_data_from_path(path="../data/data_raw/")
```

# split and save

```python
for name, df in data.items():
    if 'month' in df.columns:
        df = df[df['month'] > '2020-06']

    merged = split.merge(df)
    train = merged[merged['validation']==0].reset_index(drop=True)
    val = merged[merged['validation']==1].reset_index(drop=True)
    test = merged[merged['validation'].isna()].reset_index(drop=True)
    
    
    train[df.columns].to_csv(f"../data/data_split/train/{name}.csv", index=False)
    val[df.columns].to_csv(f"../data/data_split/val/{name}.csv", index=False)
    test[df.columns].to_csv(f"../data/data_split/test/{name}.csv", index=False)
```
